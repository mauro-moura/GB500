```python
from scikit_posthocs import posthoc_dunn
from scipy.stats import kruskal
import pandas as pd

#df = pd.read_excel('tabela_kruskall.xlsx')
df = pd.read_csv('Kruskal.csv')
first_loc = 'Dice'
sec_group = 'Arq'

data = [df.loc[ids, first_loc].values for ids in df.groupby(sec_group).groups.values()]

kw = kruskal(*data)
dunn = posthoc_dunn(df, val_col = first_loc, group_col = sec_group, p_adjust = 'holm')

titulo = '50_epocas'
kw = pd.DataFrame(kw, columns = [titulo], index = ['X²', 'p-value'])
kw.to_excel('Kruskall_Wallis%s.xlsx'%(titulo))

dunn = pd.DataFrame(dunn)
dunn.to_excel('Dunn%s.xlsx'%(titulo))
```
